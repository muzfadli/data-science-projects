```python
def satukandf(string):
    # List berisi seluruh dataframe sebelum disatukan menjadi satu dataframe
    daftar = [eval(string+"{0}".format(i+1)) for i in range(jumlahbulan)]

    # menggabungkan semua dataframe, mengisi cell berisi "NaN" dengan 0, dan mereset indeks (karena indeks sebelumnya ada di kolom SQLDATE)
    tabel = pd.concat(daftar, sort=True)

    return tabel
```

```python
tabelA = satukandf("tabelA")
tabelB = satukandf("tabelB")
tabelB = tabelB.set_index('Tanggal')
# satukandf("tabelD")
```

```python
tabelA.columns = ['Jumlah Berita Label 0','Jumlah Berita Label 1','Jumlah Berita Label 2','Jumlah Berita Label 3','Jumlah Berita Label 4','Tanggal']
```
