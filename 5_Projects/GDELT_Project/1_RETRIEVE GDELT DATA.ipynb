# Download

Berikut adalah contoh script untuk mengunduh data GDELT dengan country code ID. Selanjutnya dari csv bisa dipindah ke mysql

```python
import os.path
import urllib
import zipfile
import glob
import operator

outfilecounter = 0

#Definisikan path di sini. Path ini akan digunakan sampai program selanjutnya
local_path = 'E:/GDELT JULI/' #tempat file yang mau diekstrak TSVnya
output = 'E:/GDELT JULI/IDN/' #folder output TSV, jangan lupa foldernya dibuat dulu

fips_country_code = 'ID'
files = glob.glob(local_path + '*')

for compressed_file in files:
    print(compressed_file)

    # extract the contents of the compressed file to a temporary directory
    print('extracting,'),
    z = zipfile.ZipFile(file=compressed_file, mode='r')
    z.extractall(path=local_path + 'tmp/')

    # parse each of the csv files in the working directory,
    print('parsing,'),
    for infile_name in glob.glob(local_path + 'tmp/*'):
        outfile_name = output + fips_country_code + '%04i.tsv' % outfilecounter

        # open the infile and outfile
        with open(infile_name, encoding="utf8", mode='r') as infile, open(outfile_name, mode='w', encoding="utf8") as outfile:
            for line in infile:
                # extract lines with our interest country code
                if fips_country_code in operator.itemgetter(51, 37, 44)(line.split('\t')):
                    outfile.write(line)
            outfilecounter += 1

        # delete the temporary file
        os.remove(infile_name)
os.rmdir(local_path+'tmp/')
print('\n====================================================\n')
print('\nFILE SUDAH SELESAI DIEKSTRAK SEMUA GAN\n')
print('\n====================================================')
```

Berikut adalah contoh untuk melakukan pickling data GDELT. Setelah menjalani program berikut, semua data TSV di folder output akan menjadi satu dataframe, bernama "DF".

```python
import pandas as pd

# Jangan lupa taruh file 'CSV.header.fieldids.xlsx' di folder yang sama dengan Jupyter Notebook dijalanin
colnames = pd.read_excel('CSV.header.fieldids.xlsx', sheet_name='Sheet1',
                         index_col='Column ID', usecols=1)['Field Name']

# Membuat dataframe dari per data TSV
files = glob.glob(output + fips_country_code + '*')
DFlist = []
for active_file in files:
    print(active_file)
    DFlist.append(pd.read_csv(active_file, sep='\t', header=None, dtype=str,
                              names=colnames, index_col=['GLOBALEVENTID']))

# Menyatukan dataframe pritilan tadi ke satu satuan dan membuat backup file. Backup file bisa ditemukan di folder yang isinya kompresan file CSV tadi. Formatnya .pickle.
DF = pd.concat(DFlist)
DF.to_pickle(local_path + 'backupsmall' + fips_country_code + '.pickle')

# once everythin is safely stored away, remove the temporary files
# for active_file in files:
#     os.remove(active_file)
```

# Scraping GDELT
Berikut adalah kode contoh untuk melakukan scraping dengan library newspaper. Skrip ini akan berjalan jika data dari GDELT telah tersimpan ke MySQL

```python
from newspaper import Article
import pymysql.cursors

# Connect to the database
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='',
                             db='news',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)

hdr = {'User-Agent': 'Mozilla/5.0'}

with connection.cursor() as cursor:
    sql = "SELECT * FROM `gdelt` WHERE (TITLE IS NULL OR TRIM(TITLE) = '')"
    cursor.execute(sql)
    print(sql)
    result = cursor.fetchall()
    for r in result:
        try :
            r["SOURCEURL"] = r["SOURCEURL"].split('\r')
            print(r["SOURCEURL"][0])
            article = Article(r["SOURCEURL"][0])
            article.download()
            article.parse()
            article.nlp()
            with connection.cursor() as cursor:
                sql = "UPDATE `gdelt` SET title=%s,content=%s WHERE GLOBALEVENTID=%s"
                cursor.execute(sql, (article.title.encode('utf-8'), article.text.encode('utf-8'), r["GLOBALEVENTID"]))
                connection.commit()
        except :
            pass
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from collections import Counter

content = run("SELECT content FROM gdelt")


with open('english_stopwords.txt', 'r') as myfile:
    english_stops=myfile.read()

lemmatized = []
    
for i in range(0,len(content)):
    tokens = word_tokenize(list(content['content'])[i])
    lower_tokens = [t.lower() for t in tokens]
    # Retain alphabetic words: alpha_only
    alpha_only = [t for t in lower_tokens if t.isalpha()]

    # Remove all stop words: no_stops
    no_stops = [t for t in alpha_only if t not in english_stops]

    # Instantiate the WordNetLemmatizer
    wordnet_lemmatizer = WordNetLemmatizer()

    # Lemmatize all tokens into a new list: lemmatized
    lemmatized.extend([wordnet_lemmatizer.lemmatize(t) for t in no_stops])

# Create the bag-of-words: bow
bow = Counter(lemmatized)

# convert bow to dataframe
hitung = pd.DataFrame.from_dict(dict(bow),orient='index',columns=['jumlah'])

# mengurutkan dari yang paling banyak disebutkan
# hitung = hitung.sort_values("jumlah", ascending=False)

# mengurutkan sesuai abjad wordnya
hitung = hitung.reset_index().sort_values("index")

# print result
hitung
```
