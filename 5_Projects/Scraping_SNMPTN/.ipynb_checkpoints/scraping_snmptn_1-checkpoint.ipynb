### Scraping data from SNMPTN using BeautifoulSoup4

Get html from https://web.snmptn.ac.id/ptn

```python
import requests
from bs4 import BeautifulSoup

URL = 'https://web.snmptn.ac.id/ptn'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

get university data

```python
def scrape_univ_snmptn(snmptn_url):
    page = requests.get(snmptn_url)

    univ = BeautifulSoup(page.content, 'html.parser')

    universities = univ.find_all('tr')
    list_univ = []
    for i in range(0, len(universities)):
        dict_university = {}
        dict_university['univId'] = universities[i].find('a')['href'][5:]
        dict_university['univ_name'] = universities[i].find('a').text
        dict_university['url'] = 'https://web.snmptn.ac.id' + universities[i].find('a')['href']
        list_univ.append(dict_university)
        
    return list_univ
```

```python
snmptn_url = 'https://web.snmptn.ac.id/ptn'
list_univ = scrape_univ_snmptn(snmptn_url)
```

get prodi data

```python
def scrape_prodi(url_ptn):
    page = requests.get(url_ptn)

    ptn_soup = BeautifulSoup(page.content, 'html.parser')

    univ_prodi = []
    url = 'https://web.snmptn.ac.id'
    try:
        # Prodi SAINTEK
        saintek = ptn_soup.find(id='cat-SAINTEK')
        prodi_saintek = saintek.find_all('tr')
        for i in range(1,len(prodi_saintek)):
            prodi = {}
            prod = prodi_saintek[i].find_all('td')
            prodi['kode_prodi'] = prod[1].text
            prodi['nama_prodi'] = prod[2].text
            prodi['url_prodi'] = url + prodi_saintek[i].find('a')['href']
            prodi['jenis'] = 'SAINTEK'
            prodi['daya_tampung'] = prod[3].text
            univ_prodi.append(prodi)
    except:
        pass

    try:
        # Prodi SOSHUM
        soshum = ptn_soup.find(id='cat-SOSHUM')
        prodi_soshum = soshum.find_all('tr')
        for j in range(1,len(prodi_soshum)):
            prodi = {}
            prod = prodi_soshum[j].find_all('td')
            prodi['kode_prodi'] = prod[1].text
            prodi['nama_prodi'] = prod[2].text
            prodi['url_prodi'] = url + prodi_soshum[j].find('a')['href']
            prodi['jenis'] = 'SOSHUM'
            prodi['daya_tampung'] = prod[3].text
            univ_prodi.append(prodi)
    except:
        pass
    
    return univ_prodi
```

```python
ptn = 'https://web.snmptn.ac.id/ptn/112'
univ_prodi = scrape_prodi(ptn)
```

get prodi detail

```python
def get_prodi_detail(url):
    page = requests.get(url)
    prodi_soup = BeautifulSoup(page.content, 'html.parser')
    # detail prodi
    detail = prodi_soup.find_all('div', class_='panel panel-info')
    daya_tampung = detail[0].find_all('td')[1].text
    jenis = detail[0].find_all('td')[3].text
    if len(detail)==4:
        try:
            portfolio = detail[1].find('td').text
        except:
            portfolio = ''
        k = 2
    else:
        portfolio = ''
        k = 1
    jur = detail[k].find_all('td')
    jurusan_pendaftar = []
    for i in range(0,len(jur)):
        jurusan_pendaftar.append(jur[i].text)

    sebaran = detail[k+1].find_all('tr')
    # tahun sebaran
    tahun_sebaran = sebaran[0].find_all('th', class_='ce')
    tahun = []
    for th in range(0,len(tahun_sebaran)):
        tahun.append(tahun_sebaran[th].text)
    # jumlah pendaftar dan diterima
    jumlah_pendaftar = []
    jumlah_diterima = []
    jumlah_siswa_jur = []
    jumlah_siswa_prov = []
    for pend in range(0,len(tahun)):
        jumlah_pendaftar.append({str(tahun[pend]):sebaran[1].find_all('td', class_='ce')[pend].text})
        jumlah_diterima.append({str(tahun[pend]):sebaran[2].find_all('td', class_='ce')[pend].text.split('(')[0]})

    for i in sebaran[3:]:
        g = i.find_all('td')
        if len(g) == 2+len(tahun):
            nama_jur = {}
            list_angka_tahun = {}
            for pend in range(0,len(tahun)):
                list_angka_tahun[str(tahun[pend])]=g[pend+2].text
                nama_jur[str(g[0].text) + ' ' + str(g[1].text)] = list_angka_tahun
            jumlah_siswa_jur.append(nama_jur)

        elif len(g) == 1+len(tahun):
            nama_prov = {}
            list_angka_th = {}
            for pend in range(0,len(tahun)):
                list_angka_th[str(tahun[pend])]=g[pend+1].text
            nama_prov[str(g[0].text)] = list_angka_th
            jumlah_siswa_prov.append(nama_prov)

    detail_prodi = {
        'daya_tampung': daya_tampung,
        'jenis': jenis,
        'jurusan_pendaftar': jurusan_pendaftar,
        'jumlah_pendaftar': jumlah_pendaftar,
        'jumlah_diterima': jumlah_diterima,
        'jumlah_diterima_berdasarkan_jur': jumlah_siswa_jur,
        'jumlah_siswa_diterima_berdasarkan_provinsi': jumlah_siswa_prov
    }
    
    return detail_prodi
```

```python
url = 'https://web.snmptn.ac.id/ptn/112/20551'
prodi_detail = get_prodi_detail(url)
```

get all

```python
snmptn_url = 'https://web.snmptn.ac.id/ptn'
list_univ = scrape_univ_snmptn(snmptn_url)

data_snmptn = []
for i in list_univ:
    univ_prodi = scrape_prodi(i['url'])
    for j in univ_prodi:
        data = {}
        data.update(i)
        data.update(j)
        data['detail_prodi'] = get_prodi_detail(j['url_prodi'])
        data_snmptn.append(data)
```
