# Visualization with Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Data

```python
import random
# GENERATE RANDOM VALUE
value = []
for i in range(250):
    value.append(random.randrange(1,100,1))

print(value)
```

### Plot Styling

```python
# Get the available styles
print(plt.style.available)
```

```python
# Plot the default matlplotlib style
plt.title('Default MatPlotLib')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(value)
plt.show()
```

```python
# Plot using ggplot style
plt.style.use('ggplot')
plt.title('ggplot style')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(value)
plt.show()
```

```python
x = []
y = []
for i in range(100):
    x.append(random.randrange(1,100,1))
    y.append(random.randrange(1,100,1))
```

```python
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.scatter(x, y)
plt.show()
```

```python
plt.style.use('ggplot')
plt.title('Bar Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.bar(x,y)
plt.show()
```

```python
plt.style.use('ggplot')
plt.title('Histogram Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.hist(value)
plt.show()
```
